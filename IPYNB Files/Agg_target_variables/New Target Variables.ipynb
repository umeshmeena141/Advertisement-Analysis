```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
inp_df = pd.read_excel('./Microsoft Ad Optimization-AI Data_Survey Aggregated Data 21062019.xlsx',sheet_name='Sheet2')
```

```python
columns= ['Q.6 Enjoyment - Very Much','Q. 11a Aided Branding - It could have been for almost anything','Q. 12 Brand Cues - Very easily',
'Q. 20 Brand Appeal - Much more appealing','Q. 32 Understanding - Very easy','Q. 33 Relevance of Information - Very relevant',
'Q. 34 Credibility of Information - Agree strongly','Q. 35 New Information - A lot of new information','Q. 36 Brand Difference - Agree strongly'
]
new_col_name = ['Enjoyment_vey_much','Aided_Branding_almost_anything','Brand_Cues_very_easily','Brand_Appeal_more_appealing',
               'Understanding_very_easily','Relevance_Information_very_relevant','Credibility_Information_agree_strongly','New_Information_alot','Brand_Difference_strongly']
```

```python
inp_df.head()
```

```python
inp_df.columns.values
```

```python
group=[]
for col in columns:
    group.append(inp_df[['Video Name',col]].groupby(['Video Name']))
    
new_target_df = pd.DataFrame(columns=new_col_name,index=group[0].groups.keys())

for key in new_target_df.index:
    for i,col in enumerate(columns):
        new_target_df.loc[key,new_col_name[i]] = group[i].first().loc[key].values[0]
```

```python
new_target_df.fillna(0,inplace=True)
new_target_df.replace('-',0,inplace=True)
```

```python
new_target_df
```

```python
target_interest_trace = inp_df[['Video Name','Interest Trace']].groupby(['Video Name'])
```

```python
def get_5_mean(arr,n):
    lower_index = n
    upper_index = n
    while lower_index >= arr.keys()[0] and arr[lower_index]>= 0.95*arr[n]:
        lower_index-=1
    while upper_index< arr.keys()[-1]+1 and arr[upper_index]>= 0.95*arr[n]:
        upper_index+=1
    low_key = arr.keys()[0]
    frames = upper_index - lower_index-1
    return arr.iloc[lower_index+1-low_key:upper_index-low_key].mean(),frames
```

```python
interest_trace = pd.DataFrame(columns=['Peak_value', 'Peak_mean_score'])

for video in target_interest_trace.groups.keys():
    video_frames =target_interest_trace.get_group(video)['Interest Trace']
    interest_trace.loc[video,'Peak_value'] = video_frames.max()
    index = np.argmax(video_frames)
    if not math.isnan(video_frames.max()):
        interest_trace.loc[video,'Peak_mean_score'],interest_trace.loc[video,'Peak_frames'] = get_5_mean(video_frames,index)
    
    
```

```python
purchase_intent_cols = [col for col in inp_df.columns.values if 'PURCHASE INTENT' in col]
purchase_intent_cols.extend(['Video Name'])
```

```python
purchase_intent_df = inp_df[purchase_intent_cols].groupby('Video Name')
```

```python
purchase_intent_agg = purchase_intent_df.first().fillna(0)
purchase_intent_agg.replace('-',0,inplace=True)
```

```python
purchase_intent_score = purchase_intent_agg.sum(axis=1)
unique_val = (purchase_intent_agg !=0).sum(axis=1)
```

```python
persuasion_cols = [col for col in inp_df.columns.values if 'Persuasion ' in col]
persuasion_cols.extend(['Video Name'])
```

```python
persuasion_df = inp_df[persuasion_cols].groupby('Video Name')
```

```python
persuasion_mean_cols = [col for col in persuasion_cols if 'Mean' in col]
persuasion_mean_cols.extend(['Video Name'])
persuasion_mean_df = inp_df[persuasion_mean_cols].groupby('Video Name')
```

```python
persuasion_agg = persuasion_mean_df.first().fillna(0)
persuasion_agg.replace('-',0,inplace=True)
```

```python
persuasion_mean_score = persuasion_agg.sum(axis=1)
```

```python
persuasion_top_cols = [col for col in persuasion_cols if 'Makes me a little more likely to consider COMPLETE BRAND NAME' in col]
persuasion_top_cols.extend(['Video Name'])
persuasion_top_df = inp_df[persuasion_top_cols].groupby('Video Name')
```

```python
persuasion_top_agg = persuasion_top_df.first().fillna(0)
persuasion_top_agg.replace('-',0,inplace=True)
```

```python
persuasion_top_score = persuasion_top_agg.sum(axis=1)
```

```python
new_target_df = pd.DataFrame(columns=['Interest_peak','Interest_mean_score','Purchase_intent','Persuasion_mean','Persuasion_very_likely'])
```

```python
for video in interest_trace.index:
    new_target_df.loc[video,'Interest_peak'] = interest_trace.loc[video,'Peak_value']
    new_target_df.loc[video,'Interest_mean_score'] = interest_trace.loc[video,'Peak_mean_score'] 
    new_target_df.loc[video,'Interest_peak_frames'] = interest_trace.loc[video,'Peak_frames']    
    
    if unique_val[video] !=0:
        new_target_df.loc[video,'Purchase_intent'] = purchase_intent_score[video]/unique_val[video]  
    else:
        new_target_df.loc[video,'Purchase_intent'] = 0
    new_target_df.loc[video,'Persuasion_mean'] = persuasion_mean_score[video]    
    new_target_df.loc[video,'Persuasion_very_likely'] = persuasion_top_score[video]    
```

```python
new_target_df.to_excel('new_targets_21062019.xlsx')
```
